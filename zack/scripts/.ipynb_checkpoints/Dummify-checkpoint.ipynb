```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../processed_data/train_full.csv')
pd.set_option('display.max_columns',None)
```

```python
data.head()
```

```python
dummy_list = ['MSSubClass','MSZoning','Street','Alley','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle','OverallQual','OverallCond','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Heating','CentralAir','Electrical','KitchenQual','Functional','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive','PoolQC','Fence','MiscFeature','SaleType','SaleCondition', 'GarageYrBlt', 'YearBuilt', 'YearRemodAdd']
```

```python
len(dummy_list)
```

```python
def make_dummy_df(df, cols) :
    for x in cols:
        dums = pd.get_dummies(df[x], drop_first=True, prefix=x)
        df = pd.concat((df, dums), axis=1)
        df.drop([x], axis=1, inplace=True)
    return df
```

```python
data = make_dummy_df(data, dummy_list)
```

```python
data.shape
```

```python
colnames = data[data.applymap(np.isreal)==False].dropna(axis=1).columns.tolist()
len(colnames)
```
