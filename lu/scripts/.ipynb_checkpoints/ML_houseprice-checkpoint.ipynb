```python
# anova test & pair wise t-test
traino.MSSubClass.unique()
traino.LandSlope.unique()
```

```python
model = ols('SalePrice ~C(LandSlope)', data = traino).fit()
anova_table = sm.stats.anova_lm(model, typ = 2)
anova_table
```

```python
test_col = 'LandSlope'
compare = "SalePrice ~C(" + test_col + ")"
model = ols(compare, data = traino).fit()
anova_table = sm.stats.anova_lm(model, typ = 2)
display(anova_table)
```

```python
def anova_test(df, test_col):
    compare = compare = "SalePrice ~C(" + test_col + ")"
    model = ols(compare, data = df).fit()
    anova_table = sm.stats.anova_lm(model, typ = 2)
    return anova_table
anova_test(traino, "LandSlope")
```

```python
anova_list = list(map(lambda x: anova_test(traino, x), cate_col_list))
for item in anova_list:
    display(item)
```
